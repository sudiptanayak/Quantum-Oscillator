```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0=-0.1
p0=0
hbar=0.008
re_co=x0/np.sqrt(2*hbar)
im_co=p0/np.sqrt(2*hbar)
N=10
psi0=coherent(N,re_co+1j*im_co)
xvec=np.linspace(-0.5,0.5,100)
gv=np.sqrt(2/hbar)
W_coherent=wigner(psi0,xvec,xvec,g=gv)

```

```python
import matplotlib as mpl
from matplotlib import cm
```

```python
xvec=np.linspace(-0.5,0.5,100)
wmap=wigner_cmap(W_coherent)
nrm = mpl.colors.Normalize(-W_coherent.max(), W_coherent.max())
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
plt1 = axes[0].contourf(xvec, xvec, W_coherent, 100, cmap=cm.RdBu, norm=nrm)
axes[0].set_title("Standard Colormap")
cb1 = fig.colorbar(plt1, ax=axes[0])
plt2 = axes[1].contourf(xvec, xvec, W_coherent, 100, cmap=wmap) 
axes[1].set_title("Wigner Colormap")
cb2 = fig.colorbar(plt2, ax=axes[1])
fig.tight_layout()
plt.savefig('C:/Users/Sudipta Nayak/Pictures/wigner_isolatedduffing_N=25_t=0.png')
plt.show()
```
